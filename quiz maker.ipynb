```python
from tkinter import*
from tkinter.ttk import Radiobutton
window=Tk()
window.title('quiz maker')
window.geometry('600x600')

a=['(B). Mumbai','(A). River Dolphin','(A). 16th century']


def click():
    score=0
    if a[0]==selected1.get():
        score=score+1
    if a[1]==selected2.get():
        score=score+1
    if a[2]==selected3.get():
        score=score+1
    entry.insert(END,score)
    button1.config(state='disable')
    
        
button1=Button(window,width=5,height=0,text='Score',font=('arial',15,'bold'),command=click)
button1.place(x=180,y=450) 
entry=Entry(window,font=('arial',15))
entry.place(x=89,y=500)



selected1=StringVar()
label1=Label(window,text="Q1.Indian's largest city by population",fg='white',bg='black',font=('arial',15))
label1.place(x=10,y=10)
rad1=Radiobutton(window,text='(A). Delhi',value='(A). Delhi',variable=selected1)
rad1.place(x=50,y=50)
                                                                                       
rad2=Radiobutton(window,text='(B). Mumbai',value='(B). Mumbai',variable=selected1)
rad2.place(x=50,y=80)
rad3=Radiobutton(window,text='(C). Chennai',value='(C). Chennai',variable=selected1)
rad3.place(x=50,y=110)


selected2=StringVar()
label2=Label(window,text="Q2.What is the name of Indian's national aquatic animal?",fg='white',bg='black',font=('arial',15))
label2.place(x=10,y=150)
rad1=Radiobutton(window,text='(A). River Dolphin',value='(A). River Dolphin',variable=selected2)
rad1.place(x=50,y=200)
rad2=Radiobutton(window,text='(B). Crocodile',value='(B). Crocodile',variable=selected2)
rad2.place(x=50,y=230)
rad3=Radiobutton(window,text='(C). Green Frog',value='(C). Green Frog',variable=selected2)
rad3.place(x=50,y=260)

selected3=StringVar()
label3=Label(window,text="Q3.In which century did Porguese merchants first land in Goa",fg='white',bg='black',font=('arial',15))
label3.place(x=10,y=290)
rad1=Radiobutton(window,text='(A). 16th century',value='(A). 16th century',variable=selected3)
rad1.place(x=50,y=330)
rad2=Radiobutton(window,text='(B). 17th century',value='(B). 17th century',variable=selected3)
rad2.place(x=50,y=360)
rad3=Radiobutton(window,text='(C). 18th century',value='(C). 18th century',variable=selected3)
rad3.place(x=50,y=390)


window.mainloop()
```
